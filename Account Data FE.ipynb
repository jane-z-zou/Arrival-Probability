```python
df = df.fillna(df.mode().iloc[0])
```

```python
from datetime import datetime

cycle = df['cycle_dt'].tolist()
payment = df['payment_due_dt'].tolist()
diff = []
for i in range(0, 899832):
    date1 = datetime.strptime(cycle[i], "%m/%d/%y")
    date2 = datetime.strptime(payment[i], "%m/%d/%y")
    diff.append(abs(date2 - date1))
df['diff_dates'] = diff
```
