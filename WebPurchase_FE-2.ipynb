```python
import pandas as pd
import numpy as np
import os, glob

filename = os.path.join(os.getcwd(), "WebPurchase", "purchase_records.txt")
df = pd.read_table(filename, header=None)
df.columns = ['ac_id','purchase_date']
```

```python
df.head(10)
```

```python
filename = os.path.join(os.getcwd(), "WebPurchase", "webvisit.txt")
df1 = pd.read_table(filename, header=None)
df1.columns = ['ac_id','webvisited_date','visited_page']
df1 = df1.replace(r'^\s*$', np.nan, regex=True)
df1.head(10)
```

```python
df.shape
```

```python
df1.shape
```

Purchase records (df) much larger, contains whether customer made any purchase during the year. 

Web visit logs (df1) the history of customers at pages, not purchases.

```python
from functools import reduce
dfs = [df,df1]
df_combo = reduce(lambda  left,right: pd.merge(left,right,on=['ac_id'],
                                               how='outer'), dfs)
```

```python
df_combo
```

```python
df_combo.dtypes
```

```python
df_combo = df_combo.sort_values(by=['ac_id'], ascending=False)
```

```python
df_combo
```

```python
df_combo = df_combo[df_combo.ac_id != 'ac_id']
```

```python
df_combo.isna().sum()
```

```python
df_combo = df_combo.fillna(df_combo.mode().iloc[0])
```

```python
df_combo
```

```python
df_combo.isna().sum()
```

```python
from datetime import datetime
```

```python
purchasedate_1 = df_combo['purchase_date'].tolist()
webvisitdate_2 = df_combo['webvisited_date'].tolist()
diff = []
for i in range(0, 10414408):
    date1 = datetime.strptime(purchasedate_1[i], "%m/%d/%y")
    date2 = datetime.strptime(webvisitdate_2[i], "%m/%d/%y")
    diff.append(abs(date2 - date1))
df_combo['diff_dates'] = diff
```

```python
df_combo
```

```python
df_combo.dtypes
```

```python
df_combo.describe()
```

```python
df_combo_ranked = df_combo.groupby('ac_id')['diff_dates'].max().reset_index()
df_combo_ranked.columns = ['ac_id', 'Max Date Diff']
df_combo_ranked
```

```python
from functools import reduce
dfs = [df_combo, df_combo_ranked]
df_combined = reduce(lambda  left,right: pd.merge(left,right,on=['ac_id'],
                                               how='outer'), dfs)
```

```python
df_combined
```

```python
df_combo_min = df_combo.groupby('ac_id')['diff_dates'].min().reset_index()
df_combo_min.columns = ['ac_id', 'Min Date Diff']
df_combo_min
```

```python
from functools import reduce
dfs = [df_combined, df_combo_min]
df_combined1 = reduce(lambda  left,right: pd.merge(left,right,on=['ac_id'],
                                               how='outer'), dfs)
df_combined1
```
