```python
import pandas as pd
import numpy as np
import os, glob

filename = os.path.join(os.getcwd(), "Account_FE.txt")
df = pd.read_table(filename)
df
```

```python
filename = os.path.join(os.getcwd(), "WebPurchase_FE.txt")
df1 = pd.read_table(filename)
df1
```

```python
df = df.iloc[: , 1:]
df
```

```python
df1 = df1.iloc[: , 1:]
df1
```

```python
from functools import reduce
dfs = [df1,df]
df_FE = reduce(lambda  left,right: pd.merge(left,right,on=['ac_id'],
                                               how='outer'), dfs)
```

```python
df_FE
```

```python
df_FE = df_FE.dropna() 
df_FE
```

```python
df_FE = df_FE.drop(['purchase_date','webvisited_date'], axis=1)
df_FE
```
